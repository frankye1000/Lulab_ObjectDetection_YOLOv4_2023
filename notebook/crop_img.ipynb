```python
import os
import re
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
img_paths = glob("../../img/raw_1s/*.png")
```

```python
len(img_paths)
```

```python
def get_img_info(path):
    '''
    '10'->'000010'
    '''
    P = re.findall('(\w+)', path)
    img_class = P[2]
    img_index = P[3].zfill(6)
    return img_class, img_index
```

```python
get_img_info(img_paths[0])
```

```python
# 舊位置(壺腹乳頭在右邊)
tags = ['a','b','c','d','e','f','g','h','i','j','q']
x1 = 160
x2 = 920
y1 = 1020
y2 = 1780

for i, path in enumerate(img_paths):
    img_name = path.split("\\")[-1]
    if(img_name.startswith('q')):
        # crop image
        # img_array = cv2.imread(path)
        # plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        # plt.show()

        img_array = cv2.imread(path)[x1:x2, y1:y2]
        
        # to grayscale
    #     img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)
    #     print(img_array)
    #     # plot
        # plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        # plt.show()
    
    
        img_class, img_index = get_img_info(path)
        print(img_class, img_index, '-- finish')
        cv2.imwrite('../../img/croped_1s/{}_{}.png'.format(img_class, img_index), img_array)
```

```python
# 新位置(壺腹乳頭在左邊)
tags = ['k','l','m','n','o','p','r','s','t','u']
x1 = 160
x2 = 920
y1 = 60
y2 = 820

for i, path in enumerate(img_paths):
    img_name = path.split("\\")[-1]
    if(img_name[0] in tags):

        img_array = cv2.imread(path)[x1:x2, y1:y2]
        # plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        # plt.show()
        img_class, img_index = get_img_info(path)
        print(img_class, img_index, '-- finish')
        cv2.imwrite('../../img/croped_1s/{}_{}.png'.format(img_class, img_index), img_array)
```
